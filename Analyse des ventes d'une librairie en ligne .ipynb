# **Introduction :**

**Analyse des ventes**

Après deux années de vente en ligne, une librairie souhaite faire le point et analyser les données fournies par son site. Nous allons non seulement opérer une analyse de certains indicateurs de ventes mais également une analyse plus ciblée sur les clients. Nous essaierons ainsi de déterminer s'il y a différentes catégories de clients et tâcherons aussi de vérifier s'il existe des liens entre le genre des clients et la catégorie de livres achetés, entre l’âge des clients et le montant total des achats, la fréquence d’achat, la taille du panier moyen et les catégories des livres achetés.

#### 1ère jointure  entre les dataframes products et transactions pour obtenir df1

```python
df1 = pd.merge(df_products, df_transactions, how = "outer", on = "id_prod" )
df1
```

```python
df1.isna().sum() # Observation de valeurs manquantes dans df1
```

#### 2ème jointure  entre les dataframes transactions et customers pour obtenir df2

```python
df2 = pd.merge(df_transactions, df_customers, how= "outer", on= "client_id")
df2
```

#### 3ème jointure  entre les nouveaux dataframes df1 et df2 pour obtenir df3

```python
df3 = pd.merge(df1, df2, how= "outer", on= 'date')
df3
```

#### Observation d'un produit qui n'était pas référencé dans le jeu de données product. Les variables prix et catégorie de celui-ci sont manquantes. La quantité de ce produit (221) coïncident avec le nombre de valeurs manquantes précédemment observées.

```python
df3[df3['id_prod_x'] == '0_2245']
```

```python
df3 ['price'] = df3['price'].fillna(round(df3['price'].mean(),2)) 
# Imputation par la moyenne des prix pour remplacer les valeurs manquantes (prix de 0_2245)
```

```python
df3['categ'] =df3['categ'].fillna(0) # Référence 0_2245 équivaut à un produit de catégorie 0 (commence par 0). 
# Remplacement des valeurs manquantes de la variable catégorie par 0
```

```python
# Changement du type de la variable catégorie(float) en intéger
df3['categ'] = df3['categ'].astype('Int64')
```

```python
df3
```

#### Fin du nettoyage du dataframe df3 pour en obtenir un nouveau sur lequel nous pourrons commencer à travailler df4

```python
# Observation des doublons
df3.loc[df3['date'].duplicated(keep=False),:]
```

```python
# Suppression des doublons
df3 = df3.drop_duplicates(['date'])
df3
```

```python
# Observation du nombre d'occurrences -1.0 dans la variable prix (39)
i=df3[df3['price']== -1.0].count()
i
```

```python
# Les lignes qui impliquent cette occurrence ne sont plus prises en compte dans le dataframe
df3 = df3.loc[df3["price"] != -1.0]
df3
```

##### Ces différentes opérations auront eu pour effet de mettre à jour les données concernant le produit 0_2245 ainsi que de supprimer les lignes de test.

```python
df3.isna().sum() # Obdervation de valeurs manquantes restantes (1)
```

```python
df3= df3.dropna() # A la vue des variables concernées, suppression de la valeur manquante 
```

```python
df3.isna().sum()
```


### Chiffre d'affaires par jour

```python
# Dataframe ne prenant en charge que le montant du chiffre d'affaire pour chaque jour

dfCaJour = df4.groupby(['date']).agg({'prix': 'sum'})
dfCaJour = dfCaJour.rename (columns = {'prix':'montant'})


dfCaJour
```

```python
dfCaJour[['montant']].agg(['min','max']) # Minimum et maximum de ces montants
```

```python
# Calcul d'une moyenne mobile
dfCaJour[ '7day_rolling_avg' ] = dfCaJour.montant.rolling( 7).mean()

display(dfCaJour.head(10)) 
```

```python
# Représentation graphique du chiffre d'affaire par jour ainsi que de son lissage par moyenne mobile
parameters = {'axes.labelsize': 25,'axes.titlesize': 35, "text.color" : "k",
          "xtick.color" : "k",
          "ytick.color" : "k", 'axes.labelcolor': 'k'}


plt.rcParams.update(parameters)

plt.figure( figsize = ( 12, 5)) 

plt.gca().xaxis.set_tick_params(labelsize = 15)
plt.gca().yaxis.set_tick_params(labelsize = 15)


  
sns.lineplot( x = 'date', 
             y = 'montant', 
             data = dfCaJour, 
             label = 'CA/jour') 
  
sns.lineplot( x = 'date', 
             y = '7day_rolling_avg', 
             data = dfCaJour, 
             label = 'Moyenne mobile')


plt.title( "Chiffre d'affaires par jour durant les deux ans", color='k' )


  
plt.xlabel('Date') 
    
plt.ylabel('Montant (€)')

# plt.savefig("CA_Jours.png",dpi=300)

plt.show()
```

```python
# Ajout d'une colonne indiquant les jours de la semaine

df4['jour'] = df4['date'].dt.day_name()
df4
```

```python
# Chiffre d'Affaires selon le jour de la semaine

df_CA_jour = df4.groupby(['jour']).agg({'prix' : 'sum'})
df_CA_jour = df_CA_jour. rename ( columns = {'prix' : "Chiffre d'affaires"})
week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
df_CA_jour = df_CA_jour.reindex(week)
df_CA_jour
```

```python
# Représentations graphiques du chiffre d'affaire selon le jour de la semaine

plt.rcParams["figure.figsize"] = (12, 10)

df_CA_jour.plot(kind='pie', subplots=True, autopct= '%1.1f%%', fontsize=18,  textprops={'size': 'x-large'}, 
                wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'}, startangle=90)

plt.axis('equal')
plt.title("Part du chiffre d'affaires selon  \n le jour de la semaine", color='k')
plt.legend( title = "Jours de la semaine")

#plt.savefig("Chiffre d'affaires selon le jour de la semaine .png",dpi=300)


df_CA_jour.plot(kind='line', subplots=False, color = 'k')
plt.title("Chiffre d'affaires selon le jour de la semaine", color= 'k', y=1.05)
plt.xlabel('JOUR', color = 'k', labelpad=20)
plt.ylabel("Chiffre d'affaire (Millions d'euros)", labelpad=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid()

#plt.savefig("Chiffre d'affaires selon le jour de la semaine.png",dpi=300)



plt.show()
```

```python
# COURBE DE lORENZ

def lorenz(variable, title):
    X = variable.values
    X = np.sort(X)
# Courbe de Lorenz
    X_lorenz = X.cumsum() / X.sum()
    X_lorenz = np.insert(X_lorenz, 0, 0)
    # X_lorenz[0], X_lorenz[-1]
    y = np.arange(X_lorenz.size)/(X_lorenz.size-1)
    lorenz = pd.DataFrame()
    lorenz['X'] = pd.Series(X_lorenz)
    lorenz['Y'] = pd.Series(y)
    sns.scatterplot(data=lorenz, x='Y', y='X', marker='x')

    # Diagonale
    a = np.arange(0,1,.01)
    x = a
    y = a
    
    # Graphique
    sns.lineplot(x=x,y=y)
    plt.xlim([0,1])
    plt.ylim([0,1])
    plt.title(title, color='k')
    plt.show()
    
    #plt.savefig("Courbe de Lorenz",dpi=300)
    
# Inégalité des chiffres d'affaires annuels de tous les clients
lorenz(df_Clients['prix'], "Inégalités des chiffres d'affaires entre tous les clients")



def gini(x):
    total = 0
    for i, xi in enumerate(x[:-1], 1):
        total += np.sum(np.abs(xi - x[i:]))
    return total / (len(x)**2 * np.mean(x))

print("L'indice de gini est de", gini(df_Clients['prix']))
```

###### Test statistique

###### Nous allons ici mettre en place des tests afin de déterminer si dans le cas de la possible détection d'une corrélation entre les variables ci dessus (âge et catégorie), il est plus pertinent d'utiliser un test paramétrique de type ANOVA ou un test non paramétrique. Pour ce, nous mettons en place une première méthode graphique (qqplot) associée à un test de normalité (plus fiable) afin de savoir si les données suivent ou non une loi normale et un test d'homogénéité des variances.

```python
# Méthode graphique pour déterminer si les données suivent une distribution normale.

sm.qqplot(df5['age'], line = "r")
plt.show()
```

###### Le graphique semble indiquer que la distribution ne suit pas la loi normale (les points devraient être plus alignés sur la droite). Confirmons cela par un test de normalité.

```python
# Test (de normalité) Kolmogorov-Smirnov

st.kstest(df5['age'], df5['categorie'])
```

######  Deux hypothéses :

H0 (nulle) : la distribution est gaussienne.

H1 (alternative) : La distribution est non gaussienne.

Si la pvalue du test Kolmogorov-Smirnov est inférieur à 5%, on rejette H0 et on conclue que la distribution est non gaussienne.
La pvalue = 0 ≤ 0.05, dès lors il n'y a pas de distribution normale.

```python
# Test (d'homogénéité des variances) de Levene

st.levene(df5['age'],df5['categorie'])
```

###### Deux hypothèses

H0 (nulle) : il y a homogéneité des variances (Homoscédasticité)

H1 (alternative): les variances ne sont pas égales

Le test de Levene est statistiquement significatif, l'hypothèse d'homogénéité des variances doit être rejetée. Les variances ne sont pas égales.  (Valeur-p = 0 ≤ 0.05 donc rejet de H0)

###### Dès lors que les tests de normalité et d'homogénéité des variances ne sont pas concluants nous optons pour un test non paramétrique, celui de Kruskal Wallis.

```python
# Test de Kruskal Wallis

st.kruskal(df5['age'], df5['categorie'])
```

###### La valeur-p équivaut ici à 0 (donc ≤ 0.05), ce qui nous permet de rejeter l'hypothése nulle (H0) selon laquelle il y aurait indépendance entre les deux variables au profit de l'hypothése alternative (H1) qui nous permet de supposer un lien de dépendance statistiquement significatif entre les deux variables que sont l'âge et la catégorie de livres achetés.
